```python
import pandas as pd
import numpy as np

import seaborn as sns
df=pd.read_csv("BTech.csv")
df.head()
df.info()
import matplotlib.pyplot as plt
sns.set_theme(style="darkgrid")
fig, axs = plt.subplots(2, 2, figsize=(14, 10))  # 2x2 grid

#  1. Internships vs Placement
axp=sns.countplot(data=df, x='Internships', hue='Placement', ax=axs[0, 0])
axs[0, 0].set_title("Internships vs Placement")
axp.bar_label(axp.containers[0])
#  2. DSA Done vs Placement
axp=sns.countplot(data=df, x='DSA_Done', hue='Placement', ax=axs[0, 1])
axs[0, 1].set_title("DSA vs Placement")
axp.bar_label(axp.containers[0])
#  3. Resume Ready vs Placement
axp=sns.countplot(data=df, x='Resume_Ready', hue='Placement', ax=axs[1, 0])
axs[1, 0].set_title("Resume vs Placement")
axp.bar_label(axp.containers[0])

#  4. Mistakes distribution (only unplaced)
unplaced = df[df["Placement"] == "No"]
axp=sns.countplot(data=unplaced, y='Mistakes', order=unplaced['Mistakes'].value_counts().index, ax=axs[1, 1])
axs[1, 1].set_title("Common Mistakes by Unplaced Students")

# Adjust layout
plt.tight_layout()
axp.bar_label(axp.containers[0])
plt.show()
# print(df.isnull().sum())
# ax=sns.countplot(data=d,x='DSA_Done')
# plt.title('Placement Count (Yes/No)')
# ax=sns.countplot(data=df,x='DSA_Done',hue='Placement')
# plt.legend(loc='upper right', prop={'size': 10})
# plt.title('DSA VS Placement')
# ax.bar_label(ax.containers[0])
# plt.show()
# ax=sns.countplot(data=df,x='Internships',hue='Placement')
# plt.title('Intership vs palcement')
# ax.bar_label(ax.containers[0])
# plt.show()
# print("Total Students:", len(d))
# print("Placed:", d['Placement'].value_counts().get('Yes', 0))
# print("Not Placed:", d['Placement'].value_counts().get('No', 0))
# print("Did DSA:", d['DSA_Done'].value_counts().get('Yes', 0))

# data=pd.series('CGPA_1','CGPA_2','CGPA_3','CGPA_4')
# cgpa_by_year = d.groupby('Year')['arr'].mean().reset_index()
# sns.lineplot(data=cgpa_by_year, x='Year', y='CGPA_1', marker='o')
# plt.title('Year-wise Average CGPA')




```
