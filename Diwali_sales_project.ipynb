```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
import seaborn as sns
df=pd.read_csv("Diwali Sales Data.csv",encoding = 'unicode_escape')
df.shape

df.head(10)
```

```python
# for all data information
df.info()
```

```python
# remove null colums 
df.drop(['Status','unnamed1'], axis=1, inplace=True, errors='ignore')
df.info()
```

Remove all null values in given data set 

```python
pd.isnull(df).sum()
```

```python
# drop null values
df.dropna(inplace=True)
df.shape
```

```python
# change data type float to int 
df['Amount']=df['Amount'].astype('int') 
```

```python
# Exploring Data
ax=sns.countplot(x='Gender',data=df)
for bars in ax.containers:
 ax.bar_label(bars)
plt.savefig("bar1.png")
```

```python
The graph shows that the number of Females (7842) is more than twice the number of Males (3409) in the dataset.
This indicates that female customers contribute significantly more than male customers.
```

```python
#
sales_gen=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False) 
sns.barplot(x='Gender',y='Amount',data=sales_gen)
plt.savefig("barplot.png")
```

```python
#  basis on age
ax=sns.countplot(x='Age Group',data=df, hue='Gender')
for bars in ax.containers:
 ax.bar_label(bars)
plt.savefig("3.png")
```

The majority of customers fall in the 26–35 age group, with females significantly outnumbering males across all age groups. Overall, females dominate in every segment, especially in the 26–35 and 36–45 groups.

```python
sales_age=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False) 
sns.barplot(x='Age Group',y='Amount',data=sales_age)
plt.savefig("bar.png")
```

The 26–35 age group contributes the highest purchase amount, followed by 36–45 and 18–25 groups. Spending decreases steadily in older and younger age groups, with the 0–17 group contributing the least.

```python
# orders in State 
sales_state=df.groupby(['State'],as_index=False)['Orders'].sum().sort_values( by='Orders', ascending=False) 
sns.set(rc={'figure.figsize':(25,8)})
sns.barplot(x='State',y='Orders',data=sales_state)
plt.savefig("4.png")
```

Uttar Pradesh records the highest number of orders, followed by Maharashtra, Karnataka, and Delhi. States like Telangana, Punjab, and Rajasthan show the lowest order volumes.

```python
# Amount in state
sales_state=df.groupby(['State'],as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False) 
sns.set(rc={'figure.figsize':(25,8)})
sns.barplot(x='State',y='Amount',data=sales_state)
plt.savefig("5.png")
```

Uttar Pradesh generates the highest sales amount, followed by Maharashtra, Karnataka, and Delhi. The lowest sales contributions come from Telangana, Punjab, and Rajasthan.

```python
# marital status
ax=sns.countplot(data =df ,x='Marital_Status',hue='Marital_Status')
sns.set(rc={'figure.figsize':(7,4)})
for bar in ax.containers:
    ax.bar_label(bars)
    plt.savefig("6.png")
```

The majority of customers are unmarried (0), while a smaller portion are married (1). This indicates unmarried individuals dominate the dataset.

```python
sales_state=df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False) 
sns.set(rc={'figure.figsize':(7,4)})
sns.barplot(x='Marital_Status',y='Amount',data=sales_state,hue='Gender')
plt.savefig("7.png")
```

Unmarried females contribute the highest spending, followed by married females. In both marital groups, females spend more than males.

```python
# occupation
sales_state=df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False) 
sns.set(rc={'figure.figsize':(27,8)})
ax=sns.barplot(x='Occupation',y='Amount',data=sales_state,palette="Set1",hue='Occupation')
for bar in ax.containers:
    ax.bar_label(bars)
    plt.savefig("8.png")
```

The IT sector leads in spending, followed by Healthcare and Aviation. Occupations like Agriculture, Textile, and Construction contribute the least to overall spending.

```python
#Product category
sales_state=df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values( by='Amount', ascending=False) 
sns.set(rc={'figure.figsize':(28,8)})
sns.barplot(x='Product_Category',y='Amount',data=sales_state,palette="Set1",hue='Product_Category')
# for bar in ax.containers:
#     ax.bar_label(bars)
plt.savefig("9.png")
```

Food is the top-selling category, followed by Clothing & Apparel, Electronics & Gadgets, and Footwear & Shoes. Categories like Office, Veterinary, and Hand & Power Tools contribute the least to sales.

```python
#Product category
sales_state=df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values( by='Orders', ascending=False).head(10) 
sns.set(rc={'figure.figsize':(28,8)})
ax=sns.barplot(x='Product_ID',y='Orders',data=sales_state,palette="Set1")
for bar in ax.containers:
    ax.bar_label(bar)
    plt.savefig("10.png")
```

Conclusion : Most purchases come from unmarried females aged 26–35, with the IT sector and Food category contributing the highest sales. Uttar Pradesh leads in both order volume and sales amount, while lower contributions come from smaller states and less popular product categories.
